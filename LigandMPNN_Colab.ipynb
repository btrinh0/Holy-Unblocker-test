````python
import glob
import os

# --- CONFIGURATION ---
# This points to the folder where you just generated the 100 sequences
# Make sure this matches the output_folder from your previous step!
target_folder = "./OROV_Round2_Sequences"
search_pattern = os.path.join(target_folder, "*.fa")

# --- DEFINITIONS ---
# "Sticky" Hydrophobics (The Glue)
hydrophobic = set("FILVWY")
# "Soluble" Charged (The Shell)
charged     = set("DEKR")
# "Floppy" Glycines (Avoid too many)
tiny        = set("G")

def analyze_sequence(seq):
    L = len(seq)
    if L == 0: return None

    n_hyd = sum(1 for aa in seq if aa in hydrophobic)
    n_chg = sum(1 for aa in seq if aa in charged)
    n_gly = sum(1 for aa in seq if aa in tiny)

    hyd_pct = (n_hyd / L) * 100
    chg_pct = (n_chg / L) * 100

    # Heuristic Score:
    # We want High Hydrophobicity, but we penalize if it gets > 45% (Aggregates)
    # We penalize if Glycine is too high (Floppy)
    score = n_hyd - (n_gly * 1.5)

    return {
        "seq": seq,
        "hyd_pct": hyd_pct,
        "chg_pct": chg_pct,
        "score": score
    }

# --- EXECUTION ---
files = glob.glob(search_pattern)
candidates = []

print(f"🕵️‍♂️ Scouting for binders in {target_folder}...")

if not files:
    print("❌ No .fa files found! Check if the folder path is correct.")
elif len(files) == 0:
    print(f"❌ No .fa files found in {target_folder}.")
else:
    for f in files:
        with open(f, 'r') as file:
            lines = file.readlines()
            # LigandMPNN .fa files usually have:
            # >Structure_name, score=...
            # SEQUENCE

            # We skip the first 2 lines (Original Input) and look at the generated ones
            for i in range(2, len(lines), 2):
                if i+1 >= len(lines): break

                header = lines[i].strip()
                # Parse sequence: remove any slashed parts if present (like AAAAA:BBBB)
                full_seq_line = lines[i+1].strip()

                # If the sequence contains a colon ':', it means it has multiple chains.
                # We only want the designed chain (usually the first part or specific part).
                # Assuming standard MPNN output where the designed seq is the whole line
                # or split by colon. Let's take the longest segment or the first one.
                parts = full_seq_line.split(":")
                seq = max(parts, key=len) # Grab the longest chain (usually the binder in this context)

                stats = analyze_sequence(seq)
                if stats:
                    stats["name"] = header
                    stats["file"] = os.path.basename(f)
                    candidates.append(stats)

    # SORTING LOGIC:
    # We want the "Stickiest" ones that aren't rocks.
    # Sort by Score (High Hydrophobicity)
    candidates.sort(key=lambda x: x["score"], reverse=True)

    print(f"✅ Analyzed {len(candidates)} sequences.\n")
    print("🏆 TOP 3 CANDIDATES (Run these in AlphaFold!) 🏆")
    print("-" * 70)

    for i, c in enumerate(candidates[:3]):
        print(f"RANK #{i+1}")
        print(f"File: {c['file']}")
        print(f"Name: {c['name']}")
        print(f"Sequence: {c['seq']}")
        print(f"stats: Hydrophobic: {c['hyd_pct']:.1f}% | Charged: {c['chg_pct']:.1f}%")

        if c['hyd_pct'] < 25:
            print("⚠️ Note: Low hydrophobicity. Might be too soluble (weak binding).")
        elif c['hyd_pct'] > 35:
            print("🔥 Note: High hydrophobicity. Very sticky (Good for binding, check solubility).")
        else:
            print("✨ Note: Goldilocks Zone (Perfect balance).")

        print("-" * 70)
````

````python
import glob
import os

# --- CONFIGURATION ---
target_folder = "./OROV_Guided_Missile_Seq"
search_pattern = os.path.join(target_folder, "*.fa")

# --- DEFINITIONS ---
hydrophobic = set("FILVWY")
charged     = set("DEKR")
tiny        = set("G")

def analyze_sequence(seq):
    L = len(seq)
    if L == 0: return None

    n_hyd = sum(1 for aa in seq if aa in hydrophobic)
    n_chg = sum(1 for aa in seq if aa in charged)
    n_gly = sum(1 for aa in seq if aa in tiny)

    hyd_pct = (n_hyd / L) * 100
    chg_pct = (n_chg / L) * 100

    # Heuristic Score
    score = n_hyd - (n_gly * 1.5)

    return {
        "seq": seq,
        "hyd_pct": hyd_pct,
        "chg_pct": chg_pct,
        "score": score
    }

files = glob.glob(search_pattern)
candidates = []

print(f"🕵️‍♂️ Scouting for BINDERS (Small Chains) in {target_folder}...")

if not files:
    print("❌ No .fa files found!")
else:
    for f in files:
        with open(f, 'r') as file:
            lines = file.readlines()
            for i in range(2, len(lines), 2):
                if i+1 >= len(lines): break

                header = lines[i].strip()
                full_seq_line = lines[i+1].strip()

                # SPLIT CHAINS
                parts = full_seq_line.split(":")

                # CRITICAL FIX: Pick the chain that is Binder-sized (e.g., < 150 residues)
                # If there are multiple small ones, pick the one that isn't empty.
                binder_seq = ""
                for p in parts:
                    if 20 < len(p) < 200: # Assuming binder is between 20 and 200 AA
                        binder_seq = p
                        break

                # Fallback if logic fails (just take smallest)
                if not binder_seq:
                    binder_seq = min(parts, key=len)

                stats = analyze_sequence(binder_seq)
                if stats:
                    stats["name"] = header
                    stats["file"] = os.path.basename(f)
                    candidates.append(stats)

    # Sort by "Stickiness"
    candidates.sort(key=lambda x: x["score"], reverse=True)

    print(f"✅ Analyzed {len(candidates)} sequences.\n")
    print("🏆 TOP 3 REAL CANDIDATES (Run these in AlphaFold!) 🏆")
    print("-" * 70)

    for i, c in enumerate(candidates[:3]):
        print(f"RANK #{i+1}")
        print(f"File: {c['file']}")
        print(f"Name: {c['name']}")
        print(f"Sequence: {c['seq']}")
        print(f"Length: {len(c['seq'])} residues")
        print(f"stats: Hydrophobic: {c['hyd_pct']:.1f}% | Charged: {c['chg_pct']:.1f}%")

        if c['hyd_pct'] < 28:
            print("⚠️ Note: Slightly low hydrophobicity (Risk: Weak binding)")
        elif c['hyd_pct'] > 35:
            print("🔥 Note: High hydrophobicity (Strong binding, check solubility)")
        else:
            print("✨ Note: Goldilocks Zone (Likely good).")
        print("-" * 70)
````
